```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
import pandas as pd

# Sample data (replace this with your own single row of data)
data = pd.read_csv("C:/Users/anilp/Project_Files/BVP.csv")

# Sample data
bvp = data['64.000000']

# Create a boxplot
plt.boxplot(data, bootstrap=1000)
plt.title('Boxplot of Data')
plt.show()

# Calculate variability measures
range_value = np.ptp(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
variance = np.var(data)
std_deviation = np.std(data)
mean = np.mean(data)

print(f'Range: {range_value}')
print(f'IQR: {iqr}')
print(f'Variance: {variance}')
print(f'Standard Deviation: {std_deviation}')
print(f'Mean: {mean}')

# Q-Q plot for HR data
stats.probplot(bvp, dist="norm", plot=plt)
plt.title("Q-Q Plot for BVP")
plt.show()


# # Bootstrapping
# num_bootstraps = 100000
# bootstrap_means = []

# # QQ plot
# sm.qqplot(data, line='s')
# plt.title("QQ Plot")
# plt.show()

# for _ in range(num_bootstraps):
#     # Generate a bootstrap sample by sampling with replacement
#     bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
#     bootstrap_mean = np.mean(bootstrap_sample)
#     bootstrap_means.append(bootstrap_mean)

# # Plot the bootstrapped means
# plt.hist(bootstrap_means, bins=30, edgecolor='black')
# plt.title("Bootstrapped Means Distribution")
# plt.xlabel("Mean")
# plt.ylabel("Frequency")
# plt.show()
```
